```python
%load_ext google.cloud.bigquery
```

```python
%%bigquery
SELECT
    source_year AS year,
    COUNT(is_male) AS birth_count
FROM `bigquery-public-data.samples.natality`
GROUP BY year
ORDER BY year DESC
LIMIT 15
```

```python
%%bigquery total_births
SELECT
    state AS state,
    COUNT(*) AS birth_count 
FROM `bigquery-public-data.samples.natality` where state IS NOT null
GROUP BY state 
ORDER BY birth_count DESC
LIMIT 6
```

```python
%matplotlib inline
```

```python
total_births.plot(kind='bar', x='state', y='birth_count');
```

```python
%%bigquery total_births
SELECT
    mother_age,
    COUNT(*) AS mother_count 
FROM `bigquery-public-data.samples.natality` where source_year>=1998 AND source_year<=2008
GROUP BY mother_age
```

```python
total_births.plot(kind='bar', x='mother_age', y='mother_count');
```

```python
%%bigquery total_births
SELECT
    source_year as year,
    AVG(father_age) as average_father_age_yearwise 
FROM `bigquery-public-data.samples.natality` 
WHERE source_year>=1998 AND source_year<=2008
GROUP BY source_year
ORDER BY source_year
```

```python
total_births.plot(kind='bar', x='year', y='average_father_age_yearwise');
```
